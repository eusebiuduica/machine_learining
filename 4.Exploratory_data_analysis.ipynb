## 4) Corela탵ie (Pearson / Spearman)

### 九 Exemplu  matrice de corela탵ie + heatmap (Pearson)

```python
num_cols = ["v칙rst캒","칥n캒l탵ime_cm","greutate_kg","venit"]
corr_pearson = df[num_cols].corr(method="pearson") # Se calculeaz캒 corela탵ia
plt.figure(figsize=(6,4))
sns.heatmap(corr_pearson, annot=True, fmt=".2f", cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Corela탵ie (Pearson)  variabile numerice")
plt.tight_layout()
plt.show()
```

### 游닇 Exerci탵iu  Spearman + perechi specifice
- Calculeaz캒 matricea **Spearman** 탳i deseneaz캒 heatmap.  
- F캒 un scatter cu regresie pentru `칥n캒l탵ime_cm` vs `greutate_kg`.  
- F캒 un scatter simplu pentru o pereche cu rela탵ie slab캒 (ex.: `v칙rst캒` vs `칥n캒l탵ime_cm`).

```python
num_cols = ["v칙rst캒","칥n캒l탵ime_cm","greutate_kg","venit"]
corr_spearman = df[num_cols].corr(method="spearman")
plt.figure(figsize=(6,4))
sns.heatmap(corr_spearman, annot=True, fmt=".2f", cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Corela탵ie (Spearman)  variabile numerice")
plt.tight_layout()
plt.show()
```

```python
plt.scatter(df['칥n캒l탵ime_cm'], df['greutate_kg'])
plt.xlabel('칉n캒l탵ime (cm)')
plt.ylabel('Greutate (kg)')
plt.title('Rela탵ia dintre 칥n캒l탵ime 탳i greutate')
plt.show()
```

```python
plt.scatter(df['칥n캒l탵ime_cm'], df['v칙rst캒'])
plt.xlabel('칉n캒l탵ime (cm)')
plt.ylabel('v칙rst캒 (ani)')
plt.title('Rela탵ia dintre 칥n캒l탵ime 탳i v칙rst캒')
plt.show()
```

### 九 Exemplu  heatmap din pivot (medie venit pe ora탳 칑 segment)

```python
pivot_mean = df.pivot_table(
    index="ora탳",          # fiecare r칙nd = un ora탳
    columns="segment",     # fiecare coloan캒 = un segment (A, B, C)
    values="venit",        # valorile analizate = venitul
    aggfunc="mean"         # agregare = media veniturilor
)

plt.figure(figsize=(5,3.8))
sns.heatmap(pivot_mean, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Media venit  ora탳 칑 segment")
plt.tight_layout()
plt.show()
```

### 游닇 Exerci탵iu
- Construie탳te un pivot cu **mediana** `venit` pe `ora탳` 칑 `gen`.  
- Deseneaz캒 heatmap 탳i interpreteaz캒 diferen탵ele.

```python
pivot_mean = df.pivot_table(
    index="ora탳",
    columns="gen",
    values="venit",
    aggfunc="median"
)

plt.figure(figsize=(5,3.8))
sns.heatmap(pivot_mean, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Mediana venit  ora탳 칑 gen")
plt.tight_layout()
plt.show()
```

Din heatmap se observa ca M are un venit median mai mare in fiecare oras.
De asemenea, in Bucuresti venitul median este cel mai mare, iar in Timisoara cel mai mic.

### Vizualiz캒ri categorice

```python
plt.figure(figsize=(15, 4))

# 1) Boxplot  rezumat simplu (median캒, IQR, outlieri)
plt.subplot(1, 3, 1)
sns.boxplot(data=df, x="segment", y="venit", hue=None)          # sau hue="gen_clean"
plt.title("Boxplot  Venit per Segment")
plt.xlabel("Segment"); plt.ylabel("Venit")

# 2) Violinplot  form캒 + distribu탵ie
plt.subplot(1, 3, 2)
sns.violinplot(data=df, x="segment", y="venit", inner="quartile")
plt.title("Violinplot  Venit per Segment")
plt.xlabel("Segment"); plt.ylabel("Venit")

# 3) Swarmplot  fiecare client vizibil (aten탵ie: poate fi lent cu multe r칙nduri)
plt.subplot(1, 3, 3)
sns.swarmplot(data=df, x="segment", y="venit", size=2, alpha=0.6)
plt.title("Swarmplot  Venit per Segment")
plt.xlabel("Segment"); plt.ylabel("Venit")

plt.tight_layout()
plt.show()
```

### Plot-uri interactive

```python
import plotly.express as px

# 較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較
# 1) Scatter interactiv: v칙rst캒 vs venit
fig = px.scatter(
    df,
    x="v칙rst캒",
    y="venit",
    color="gen",  # f/m (sau schimb캒 칥n "segment")
    hover_data=["segment", "ora탳", "칥n캒l탵ime_cm", "greutate_kg", "data_inscriere"],
    title="Clien탵i: v칙rst캒 vs venit (interactiv)"
)
fig.update_traces(marker=dict(size=7, opacity=0.7))
fig.update_layout(legend_title_text="Gen")
fig.show()
```

```python
# 較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較
# 2) Histogram캒 interactiv캒: venit (pe gen)
fig = px.histogram(
    df,
    x="venit",
    nbins=20,
    color="gen",       # scoate dac캒 vrei o singur캒 culoare
    barmode="overlay",       # suprapune femei/b캒rba탵i
    histnorm="probability density",  # comparabil cu KDE
    opacity=0.6,
    title="Distribu탵ia veniturilor  Histogram캒 interactiv캒"
)
fig.update_layout(bargap=0.05)
fig.show()
```

```python
# 較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較較
# 3) Boxplot interactiv: venit per segment
fig = px.box(
    df,
    x="segment",
    y="venit",
    color="segment",
    points="outliers",      # arat캒 outlierii
    hover_data=["gen", "ora탳"],
    title="Venit per segment  Boxplot interactiv"
)
fig.show()
```
