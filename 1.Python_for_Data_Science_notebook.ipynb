# üìä Pandas ‚Äì Celula 4: GroupBy »ôi agregƒÉri

De multe ori vrem sƒÉ analizƒÉm **subgrupuri** din date »ôi sƒÉ calculƒÉm statistici pe fiecare.

1. **Ce face GroupBy**
   - √émparte tabelul √Æn grupuri pe baza uneia sau mai multor coloane.
   - Exemplu: grupƒÉm pasagerii dupƒÉ `Pclass` »ôi `Sex`.

2. **Agregarea (agg)**
   - Pe fiecare grup putem calcula valori rezumative (agregate).
   - Exemple de func»õii:
     - `size` ‚Üí c√¢te r√¢nduri are grupul.
     - `mean` ‚Üí media.
     - `median` ‚Üí valoarea din mijloc.
     - `sum` ‚Üí suma valorilor.

3. **De ce e util**
   - Putem compara grupuri √Æntre ele.
   - Exemplu: rata de supravie»õuire diferƒÉ √Æntre femeile din clasa 1 »ôi bƒÉrba»õii din clasa 3.

üëâ GroupBy + agg = ‚Äûia datele, √Æmparte-le √Æn grupuri »ôi rezumƒÉ-le prin statistici‚Äù.


```python
# GroupBy pe clasa biletului »ôi sex; util: √Æn»õelegem diferen»õe pe subgrupuri
group_stats = (
    small
    .groupby(["Pclass", "Sex"])
    .agg(
        count=("Survived", "size"),       # mƒÉrimea grupului (c√¢te r√¢nduri)
        survived_rate=("Survived", "mean"), # rata de supravie»õuire pe grup
        age_median=("Age", "median"),     # mediana (valoarea de mijloc) v√¢rstei
        fare_mean=("Fare", "mean")        # media tarifelor
    )
    .reset_index()
)

# Vedem rezultatul; util: insight-uri descriptive pentru discu»õie
group_stats
```

# üå∏ Exerci»õiu: GroupBy »ôi agregƒÉri pe Iris

1. Grupa»õi florile dupƒÉ coloana `species`.
2. Calcula»õi pentru fiecare specie:
   - `count` = c√¢te flori sunt.
   - `sepal_length_mean` = media lungimii sepalei.
   - `petal_width_median` = mediana lƒÉ»õimii petalei.
3. Afi»ôa»õi tabelul rezultat.


```python
# GroupBy pe specie; calculƒÉm statistici pentru fiecare grup
iris_stats = (
    iris
    .groupby("species")
    .agg(
        count=("species", "size"),                # c√¢te flori per specie
        sepal_length_mean=("sepal_length", "mean"), # media sepal_length
        petal_width_median=("petal_width", "median") # mediana petal_width
    )
    .reset_index()
)

iris_stats
```

# üß± Pandas ‚Äì Celula 5: Feature engineering (coloanƒÉ binarƒÉ + `pd.qcut`)

√én aceastƒÉ celulƒÉ adƒÉugƒÉm **caracteristici noi** (features) care ajutƒÉ analiza/modelarea.

1) **ColoanƒÉ binarƒÉ (`IsChild`)**
- Ideea: transformƒÉm o regulƒÉ simplƒÉ de business √Æntr-o variabilƒÉ 0/1.
- Exemplu: ‚Äûeste copil?‚Äù ‚Üí `Age < 12` ‚áí `1`, altfel `0`.
- De ce e util: uneori regulile logice surprind bine un fenomen (ex. copiii au avut alte prioritƒÉ»õi la √Æmbarcare/supravie»õuire).

2) **Discretizare cu `pd.qcut` (Fare ‚Üí FareBand)**
- `qcut` = √Æmparte o coloanƒÉ numericƒÉ √Æn **intervale** care con»õin **aproximativ acela»ôi numƒÉr de r√¢nduri**.
- `q=4` √ÆnseamnƒÉ 4 ‚Äûquartile‚Äù (fiecare ~25% din date).
- DƒÉm etichete prietenoase: `"low"`, `"mid-low"`, `"mid-high"`, `"high"`.
- De ce e util: √Æn loc de un numƒÉr brut (ex. `Fare=128.0`), putem lucra cu **categorii** (‚Äûbilet ieftin / scump‚Äù), u»ôor de explicat »ôi de comparat.

3) **Verificare distribu»õie**
- `value_counts()` aratƒÉ c√¢te r√¢nduri intrƒÉ √Æn fiecare bandƒÉ.
- Ar trebui sƒÉ fie relativ echilibrate (pentru cƒÉ `qcut` √ÆncearcƒÉ echilibru pe numƒÉr de r√¢nduri, nu neapƒÉrat pe interval numeric).


```python
small["IsChild"] = (small["Age"] < 12).astype(int)
small.head(10)
```

```python
# Cream o coloanƒÉ binarƒÉ "IsChild" (copii sub 12 ani); util: feature cu semnifica»õie realƒÉ
small["IsChild"] = (small["Age"] < 12).astype(int)

# Cream o coloanƒÉ categoricƒÉ "FareBand" pentru a discretiza Fare √Æn intervale; util: analizƒÉ pe categorii de pre»õ
small["FareBand"] = pd.qcut(small["Fare"], q=5, labels=["low", "mid-low", "mid", "mid-high", "high"])
small.head()
# VerificƒÉm distribu»õia dupƒÉ FareBand; util: echilibru pe quartile
small["FareBand"].value_counts()
```

```python
small["FareBand"] = pd.cut(small["Fare"], bins=4, labels=["low", "mid-low", "mid-high", "high"])
small["FareBand"].value_counts()
```

# üå∏ Exerci»õiu: Feature engineering pe Iris

1) AdƒÉuga»õi o coloanƒÉ binarƒÉ `IsWideSepal`:
   - `IsWideSepal = 1` dacƒÉ `sepal_width > 3.0`, altfel `0`.

2) Discretiza»õi `petal_length` √Æn **4 benzi** (quartile) folosind `pd.qcut`:
   - Folosi»õi etichetele: `["short", "mid-short", "mid-long", "long"]`.
   - Salva»õi rezultatul √Æntr-o coloanƒÉ `PetalLenBand`.

3) Afi»ôa»õi distribu»õiile:
   - `value_counts()` pentru `IsWideSepal`
   - `value_counts()` pentru `PetalLenBand`


```python
# 1) ColoanƒÉ binarƒÉ: este sepal_width mai mare de 3.0?
iris["IsWideSepal"] = (iris["sepal_width"] > 3.0).astype(int)

# 2) Discretizare petal_length √Æn 4 benzi cu qcut (quartile)
iris["PetalLenBand"] = pd.qcut(
    iris["petal_length"],
    q=4,
    labels=["short", "mid-short", "mid-long", "long"]
)

# 3) Distribu»õii
print("Distribu»õie IsWideSepal (0/1):")
print(iris["IsWideSepal"].value_counts(), "\n")

print("Distribu»õie PetalLenBand (quartile):")
print(iris["PetalLenBand"].value_counts())
```

# üîÑ Pandas ‚Äì Celula 6: Tabel pivot

## 1. Ce este un tabel pivot?
- Un **tabel pivot** este un mod de a **reorganiza »ôi rezuma datele** √Æntr-o matrice (r√¢nduri √ó coloane).
- Este foarte des folosit »ôi √Æn Excel (dacƒÉ nu a»õi lucrat cu el: practic vƒÉ lasƒÉ sƒÉ lua»õi un tabel mare »ôi sƒÉ-l rezuma»õi rapid dupƒÉ categorii).
- √én Pandas, `pivot_table()` face acela»ôi lucru: alegem o coloanƒÉ pe r√¢nduri (`index`), o alta coloanƒÉ pentru coloanele tabelului mai mic (`columns`), »ôi o **func»õie de agregare** (`aggfunc`) care calculeazƒÉ valori √Æn interiorul matricei.

## 2. Diferen»õa fa»õƒÉ de GroupBy
- `groupby()` ‚Üí produce un tabel ‚Äûlung‚Äù, unde fiecare combina»õie de categorii este un r√¢nd.  
- `pivot_table()` ‚Üí aranjeazƒÉ acelea»ôi rezultate √Æntr-o matrice mai u»ôor de citit »ôi comparat (categorii pe r√¢nduri »ôi coloane).

üëâ Acela»ôi con»õinut, doar aranjat diferit pentru prezentare mai clarƒÉ.


```python
# PivotƒÉm rata de supravie»õuire pe Pclass (r√¢nd) »ôi Sex (coloanƒÉ); util: tabel clar pentru prezentare
pivot_survival = small.pivot_table(
    values="Survived", # vrem sƒÉ analizƒÉm coloana Survived.
    index="Pclass", # fiecare r√¢nd = o clasƒÉ de bilet. Pclass reprezinta cele 3 clase de bilet posibile
    columns="Sex", # fiecare coloana va fi atributa unui Sex
    aggfunc="mean" # √Æn fiecare celulƒÉ, calculƒÉm media supravie»õuirii.
)

# Afi»ôƒÉm pivotul; logic: comparƒÉm foarte rapid grupuri principale
pivot_survival
```

# üìà Matplotlib ‚Äì Histograme

Este o biblioteca python pentru vizualizarea datelor.

## 1. Ce este o histogramƒÉ?
- O histogramƒÉ aratƒÉ **cum sunt distribuite valorile** unei coloane numerice.
- Intervalele (numite *bins*) √Æmpart scala variabilei √Æn bucƒÉ»õi, iar fiecare barƒÉ aratƒÉ c√¢te valori cad √Æn acel interval.

- `edgecolor`: culoarea marginilor barelor (ajutƒÉ la lizibilitate).
- `title`, `xlabel`, `ylabel`: context pentru cititor.

## 3. C√¢nd folosim histograma?
- C√¢nd vrem sƒÉ vedem **distribu»õia** unei variabile numerice.
- Exemple:
  - Distribu»õia v√¢rstelor √Æntr-un e»ôantion.

# üñºÔ∏è Matplotlib ‚Äì Subplots

## 1. Ce este un subplot?
- Uneori vrem sƒÉ comparƒÉm mai multe grafice **alƒÉturat** √Æn aceea»ôi figurƒÉ.
- `plt.subplots()` creeazƒÉ o figurƒÉ care con»õine o grilƒÉ de grafice (ex: 1 r√¢nd √ó 2 coloane).

## 2. Cum func»õioneazƒÉ?
- `fig, axes = plt.subplots(nrows, ncols, figsize=(width, height))`
  - `nrows`, `ncols` = c√¢te r√¢nduri »ôi coloane de grafice vrem.
  - `fig` = √Æntreaga figurƒÉ.
  - `axes` = o listƒÉ de obiecte pe care desenƒÉm fiecare plot.
- Exemplu: `axes[0]` = primul grafic, `axes[1]` = al doilea.

## 3. De ce folosim subplots?
- Ca sƒÉ comparƒÉm rapid douƒÉ sau mai multe variabile, una l√¢ngƒÉ alta.
- Exemple:
  - HistogramƒÉ pentru `Age` »ôi `Fare`.
  - Distribu»õia sepalelor »ôi petalelor √Æn Iris.


```python
# Cream o figurƒÉ cu grilƒÉ 1x2 (1 rand, 2 coloane) de subplots; util: comparƒÉm douƒÉ distribu»õii alƒÉturat
fig, axes = plt.subplots(1, 2, figsize=(12, 4.5), sharey=False)

# HistogramƒÉ pentru Age pe primul subplot (axes[0])
axes[0].hist(small["Age"].dropna(), bins=20,edgecolor="black")
axes[0].set_title("Distribu»õia Age")
axes[0].set_xlabel("Age")
axes[0].set_ylabel("Count")

# HistogramƒÉ pentru Fare pe al doilea subplot (axes[1])
axes[1].hist(small["Fare"].dropna(), bins=20, edgecolor="black")
axes[1].set_title("Distribu»õia Fare")
axes[1].set_xlabel("Fare")

# AjustƒÉm layout ca etichetele sƒÉ nu se suprapunƒÉ
plt.tight_layout()

# Afi»ôƒÉm
plt.show()
```
